```python
# Decimal to Binary, Octal, Hexadecimal


num = int(input("Enter a decimal number: "))

print("Binary:", format(num, 'b')) # Decimal → Binary
print("Octal:", format(num, 'o')) # Decimal → Octal
print("Hexadecimal:", format(num, 'X')) # Decimal → Hex


```

```python
# Binary to Decimal, Octal, Hexadecimal
binary = input("Enter a binary number: ")
decimal = int(binary, 2)
print("Decimal:", decimal)

print("Octal:", format(int(binary, 2), 'o')) #Binary to Octal
print("Hexadecimal:", format(int(binary, 2), 'X')) #Binary to Octal
```

```python
# Octal → Decimal, binary, hexadecimal
octal = input("Enter an octal number: ")
decimal = int(octal, 8)
print("Decimal:", decimal)

decimal = int(octal, 8)
print("Binary:", format(decimal, 'b')) # Decimal → Binary
print("Octal:", format(decimal, 'o')) # Decimal → Octal
print("Hexadecimal:", format(decimal, 'X')) # Decimal → Hex
```

```python
# Hexadecimal → Decimal
hexadecimal = input("Enter a hexadecimal number: ")
decimal = int(hexadecimal, 16)
print("Decimal:", decimal)

print("Binary:", format(decimal, 'b')) # Decimal → Binary
print("Octal:", format(decimal, 'o')) # Decimal → Octal
print("Hexadecimal:", format(decimal, 'X')) # Decimal → Hex
```

```python
def fraction_to_base(fraction, base, precision=10):
    result = ""
    count = 0

    while fraction > 0 and count < precision:
        fraction *= base
        digit = int(fraction)

        if digit < 10:
            result += str(digit)
        else:
            result += chr(digit - 10 + ord('A'))  # for hex

        fraction -= digit
        count += 1

    return result


num = float(input("Enter a decimal number: "))

integer_part = int(num)
fraction_part = num - integer_part

# Convert integer part normally
binary_int = format(integer_part, 'b')
octal_int = format(integer_part, 'o')
hex_int = format(integer_part, 'X')

# Convert fraction part
binary_frac = fraction_to_base(fraction_part, 2)
octal_frac = fraction_to_base(fraction_part, 8)
hex_frac = fraction_to_base(fraction_part, 16)

print("Binary:", binary_int + "." + binary_frac)
print("Octal:", octal_int + "." + octal_frac)
print("Hexadecimal:", hex_int + "." + hex_frac)
```
